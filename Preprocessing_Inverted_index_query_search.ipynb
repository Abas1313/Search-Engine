#### The class below is to pre_process the data ,
#### cleaning it by removing the unnncessriy special cases and converting upperr case to lower 

```python
class pre_processing:
    
    def remove_lower_case(Inverted_index):
        clean_data = pd.DataFrame()
        clean_data["docId"]=Inverted_index["docId"]
        stop_words = set(stopwords.words('english')) 
        clean_data["Author"] = Inverted_index["Author"].apply(lambda x: ' '.join([word.lower() for word in x.split() if word not in (stop_words)]) )
        clean_data["paper_title"] = Inverted_index["paper_title"].apply(lambda x: ' '.join([word.lower() for word in x.split() if word not in (stop_words)]) )
        
        clean_data["Author"] = clean_data["Author"].apply(lambda string: ''.join(char for char in string if char not in punctuations))
        
        clean_data["paper_title"] = clean_data["paper_title"].apply(lambda string: ''.join(char for char in string if char not in punctuations))
        
        clean_data["paper_link"]=Inverted_index["paper_link"] 
        return clean_data

        
    clean_data = remove_lower_case(Inverted_index)
    
    # text data normalisation using lemmitiser 

    
    def text_data_lemitiser( text_data):
        lemitised_data = ""
        for word in text_data.split():
            lemitise = wordnet_lemmatizer.lemmatize(word)
            lemitised_data += lemitise
            lemitised_data += " "
        lemitised_data = lemitised_data.strip()
        return lemitised_data
    
    clean_data['Author'] = clean_data['Author'].apply(text_data_lemitiser)
    clean_data['paper_title'] = clean_data['paper_title'].apply(text_data_lemitiser)
    clean_data['text_data'] = clean_data['Author'] + " " + clean_data['paper_title']
    combined_data=clean_data
    drop_columns= ['Author', 'paper_title','paper_link']
    combined_data = combined_data.drop(drop_columns, axis=1)
```

```python
class Inverted_index:
    
    # the function will enable the words 
    #within the index to be searched as a single word and for each word append 
    #
    
    def index_vacob_1(self,point_entry, index):
        words = point_entry.text_data.split()
        docId = point_entry.docId
        for word in words:
            if word in index.keys():
                if docId not in index[word]:
                    index[word].append(docId)
            else:
                index[word] = [docId]
        return index
    
    # function below will iterate overall the text data and append it to index
    def index_vacob_all(self,data, index):
        try:
            for i in range(len(data)):
                point_entry = data.loc[i,:]
                index = self.index_vacob_1(point_entry = point_entry, index = index)
        except:
            pass
        return index
    
    # the function below is build the invertedd index by calling the previous function inside it
        
    def inverted_index_builder(self,data, index):
        to_add = data
        index = self.index_vacob_all(data = to_add, index = index)
        return index
    
    # the function below is return the complete inverted index
    def t(self):
        inverted_index_database = self.inverted_index_builder(data = combined_data, index = {})
        return inverted_index_database
```

```python
class seach_query:
    
    def text_data_lemitiser(self,text_data):
        lemitised_data = ""
        for word in text_data.split():
            lemitise = wordnet_lemmatizer.lemmatize(word)
            lemitised_data += lemitise
            lemitised_data += " "
        lemitised_data = lemitised_data.strip()
        return lemitised_data
    
    def clean_query(self,text):
        text = text.lower() #to lowercase
        text = text.translate(str.maketrans('', '', string.punctuation))
        text = self.text_data_lemitiser(text)
        return text
    def process_query(self,query):
        norm = self.clean_query(query)
        return norm.split()


    
    def lists_int(self,lists):
        intersect = list(set.intersection(*map(set, lists)))
        intersect.sort()
        return intersect
    
    def my_search_engine(self,query, index=inverted_index_database):
        query_split = self.process_query(query)
        retrieved_data = []
        for word in query_split:
            if word in index.keys():
                retrieved_data.append(index[word])
        if len(retrieved_data)>0:
            result = self.lists_int(retrieved_data)
        else:
            result = [0]
        return result
    
    def t(self, user_input):
        search_result = self.my_search_engine(user_input, index=inverted_index_database)
        return search_result
```
